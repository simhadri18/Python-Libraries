# apply , map , applymap

# merge , join and concat

```python
df2=pd.DataFrame({'Fruits':['mango','grapes','banana','watermelon'],
                 'Price':[32,40,346,50]})
df2
```

```python
import pandas as pd
df1=pd.DataFrame({'Fruits':['mango','guava','banana'],
                 'Price':[30,40,35]})
df1
```

```python
#left join
merg=pd.merge(df1,df2,how='left')  #defalut it takes fruit , we can change col where u want to join on basis
merg
```

```python
#defalut it takes fruit , we can change col where u want to join on basis
merg=pd.merge(df1,df2,how='left',on='Price')  
merg
```

```python
#right join
merg=pd.merge(df1,df2,how='right',on='Fruits')  
merg
```

```python
#inner join 
merg=pd.merge(df1,df2,how='inner',on='Price')  
merg
```

```python
#outer join
merg=pd.merge(df1,df2,how='outer',on='Price')  
merg
```

```python
import pandas as pd
df=pd.DataFrame({'foo':['one','one','one','two','two','two'],
                'bar':['A','B','C','A','B','C'],
                'baz':[1,2,3,4,5,6],
                'zoo':['x','y','z','q','w','t']})
df
```

```python
df.pivot(index='foo',columns='bar',values='baz')
```

```python
df.pivot(index=['foo','zoo'],columns='bar',values='baz')
```

```python
# we pivot table instead because it allows more than one row or col   duplicates entries are allowed 
df.pivot_table(index='foo',columns='bar',values='baz')
```

```python
df.pivot_table(index=['foo','zoo'],columns='bar',values='baz')
```

# Melt

```python
import pandas as pd
df=pd.DataFrame({'foo':['one','one','one','two','two','two'],
                'bar':['A','B','C','A','B','C'],
                'baz':[1,2,3,4,5,6],
                'zoo':['x','y','z','q','w','t']})
df
```
