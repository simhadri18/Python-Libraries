keyword arguements

```python
#linewidth
import matplotlib.pyplot as plt
plt.plot([1,2,3],linewidth=20)
plt.show()
```

```python
#marker 
import matplotlib.pyplot as plt
plt.plot([1,2,3],marker=">")
plt.show()
```

```python
#marker size
import matplotlib.pyplot as plt
plt.plot([1,2,3],marker=">",markersize=9)
plt.show()
```

```python
#marker edge width 
import matplotlib.pyplot as plt
plt.plot([1,2,3],marker=">",markeredgewidth=20)
plt.show()
```

```python
#marker edge color
import matplotlib.pyplot as plt
plt.plot([1,2,3],marker=".",markeredgecolor="r")
plt.show()
```

```python
#marker face color 
import matplotlib.pyplot as plt
plt.plot([1,2,3],marker=".",markerfacecolor="r",markeredgewidth=10)
plt.show()
```

```python
#mark evry --> where we want to mark
#marker 
import matplotlib.pyplot as plt
plt.plot([1,2,3],marker="+",markevery=1)
plt.show()
```

```python
#dashes
#marker 
import matplotlib.pyplot as plt
plt.plot([1,2,3],marker=">",dashes=[4,7])
plt.show()
```

```python
#anti aiased
#marker 
import matplotlib.pyplot as plt
plt.plot([1,2,3],linewidth=20,antialiased=True)
plt.show()
```

```python
#anti aiased
#marker 
import matplotlib.pyplot as plt
plt.plot([1,2,3],linewidth=20,antialiased=False)
plt.show()
```

# zorder

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],linewidth=20,color='g')
plt.plot([1,2,5,6],linewidth=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],linewidth=20,color='g',zorder=1)
plt.plot([1,2,5,6],linewidth=20,zorder=2)
plt.show()
```

# alpha

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],linewidth=20,alpha=0.1,color="b")
plt.plot([1,2,5,6],linewidth=20,color="b")
plt.show()
```

# label

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],linewidth=20)
plt.xlabel("x label")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],linewidth=20)
plt.xlabel("x label",{"size":10,"color":"red"},labelpad=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],linewidth=20)
plt.xlabel("x label",{"size":"xx-small","color":"red"},labelpad=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],linewidth=20)
plt.xlabel("x label",{"size":10,"color":"red"},labelpad=10,backgroundcolor="yellow")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],linewidth=20)
plt.xlabel("x label",{"size":10,"color":"red"},labelpad=10,style="italic")
plt.show()
```
